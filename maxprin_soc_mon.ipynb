# Sistema

## Monopolist

Per il monopolista la situazione è la seguente:

\begin{align}
	\max_{h} &\int_{0}^{\infty}{e^{-\delta t}[p(h)-c(x)]h} ~dt \nonumber\\ 
	&\text{s.t.} 
	\begin{cases}
		\frac{dx}{dt} = G(x) - h(t) \\
		x(0) = x_0 \\
		x(t) \geq 0,~ 0\leq h(t) \leq h_{\max} 
	\end{cases}  
\end{align}

dove $G(x) = r/Kx(K-x)$, $p(h) = A - bH$ e $c(x) = c/(qx)$.

Dal maximum principle $T<\infty$ ottengo le seguenti equazioni:

\begin{align}
\label{eq:mon}
				\frac{dh}{dt} &= \frac{1}{M''(h)}\left[ \left(\delta - G'(x)\right)\left(M'(h)-c(x)\right) + c'(x)G(x) \right] \\
				\frac{dx}{dt} &= G(x) - h(t)
\end{align}

dove $M(h):=p(h)h=(A-bh)h \implies M'(h) = A-2bh$ e $M''(h)=-2b$ e quindi:

\begin{align*}
				\frac{dh}{dt} &= -\frac{1}{2b}\left[ \left(\delta - G'(x)\right)\left(A-2bh-c(x)\right) + c'(x)G(x) \right] \\
				\frac{dx}{dt} &= G(x) - h(t)
\end{align*}

## Social Planner

Per il social planner la situazione è la seguente:

\begin{align}
	\max_{h} &\int_{0}^{\infty}{e^{-\delta t}[U(h)-c(x)h]} ~dt \nonumber\\ 
	&\text{s.t.} 
	\begin{cases}
		\frac{dx}{dt} = G(x) - h(t) \\
		x(0) = x_0 \\
		x(t) \geq 0,~ 0\leq h(t) \leq h_{\max} 
	\end{cases}  
\end{align}

dove $G(x) = r/Kx(K-x)$, $p(h) = A - bH$, $U'(h) = p(h)$ e $c(x) = c/(qx)$.

Dal maximum principle $T<\infty$ ottengo le seguenti equazioni:

\begin{align}
				\frac{dh}{dt} &= \frac{1}{M''(h)}\left[ \left(\delta - G'(x)\right)\left(M'(h)-c(x)\right) + c'(x)G(x) \right] \\
				\frac{dx}{dt} &= G(x) - h(t)
\end{align}

dove $M(h):=U(h) \implies M'(h) = p(h) = A-bh$ e $M''(h)=-b$ e quindi:

\begin{align*}
				\frac{dh}{dt} &= -\frac{1}{b}\left[ \left(\delta - G'(x)\right)\left(A-bh-c(x)\right) + c'(x)G(x) \right] \\
				\frac{dx}{dt} &= G(x) - h(t)
\end{align*}

```python
class Monopolist:
    """
    All functions for monopolist
    """ 
    
    def __init__(self, A, b, ca, d, K, q, r):
        "Initialize  model parameters"
        self.A, self.b, self.ca, self.d, self.K, self.q, self.r = A, b, ca, d, K, q, r
    
    def c(self, x):
        "cost function"
        return self.ca /(self.q*x)
    
    def Dc(self, x):
        "derivative of cost function"
        return -self.ca/(self.q*x**2)
    
    def p(self, h):
        "inverse linear demand function"
        return self.A-self.b*h
    
    def g(self, x):
        "logistic growth"
        return self.r/self.K*x*(self.K-x)
    
    def Dg(self, x):
        "derivative of logistic growth"
        return self.r/self.K*(self.K-2*x)

    def Dx(self, x,h):
        "state equation for population"
        return self.g(x) - h

    def Dh(self, x, h):
        "harvest equation by subst. adjoint eq. from Hamiltonian"
        return -1/(2*self.b)*( (self.d-self.Dg(x))*(self.A-2*self.b*h-self.c(x))+self.Dc(x)*self.g(x) )

    def dY_dt(self, p, t=0):
        "state equations: returns array [dx,dh] for pop. and harvest"
        x, h = p[0], p[1]
        return np.array([self.Dx(x,h), self.Dh(x,h)])
    
```

```python
class SocialPlanner:
    """
    All functions for social planner
    """ 
    
    def __init__(self, A, b, ca, d, K, q, r):
        "Initialize  model parameters"
        self.A, self.b, self.ca, self.d, self.K, self.q, self.r = A, b, ca, d, K, q, r
    
    def c(self, x):
        "cost function"
        return self.ca /(self.q*x)
    
    def Dc(self, x):
        "derivative of cost function"
        return -self.ca/(self.q*x**2)
    
    def p(self, h):
        "inverse linear demand function"
        return self.A-self.b*h
    
    def g(self, x):
        "logistic growth"
        return self.r/self.K*x*(self.K-x)
    
    def Dg(self, x):
        "derivative of logistic growth"
        return self.r/self.K*(self.K-2*x)

    def Dx(self, x,h):
        "state equation for population"
        return self.g(x) - h

    def Dh(self, x, h):
        "harvest equation by subst. adjoint eq. from Hamiltonian"
        return -1/(self.b)*( (self.d-self.Dg(x))*(self.A-self.b*h-self.c(x))+self.Dc(x)*self.g(x) )

    def dY_dt(self, p, t=0):
        "state equations: returns array [dx,dh] for pop. and harvest"
        x, h = p[0], p[1]
        return np.array([self.Dx(x,h), self.Dh(x,h)])
    
```

```python
from __future__ import division  
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as opt


parameters = 10, 2, 1, 0.5, 15, 1, 0.6  # A, b, ca, d, K, q, r
mon = Monopolist(*parameters)
soc = SocialPlanner(*parameters)

# ============== find fixed points ================
fp_mon = []
fp_soc = []
for h in range(3):
    for x in range(15):
        try:
            fp_mon.append(opt.newton_krylov(mon.dY_dt, [x,h]))
            fp_soc.append(opt.newton_krylov(soc.dY_dt, [x,h]))
        except:
            fp_mon.append('error')
            fp_soc.append('error')

            
```

```python
# =========== phase-space plot ====================
gsize = 30
xmin, xmax = 1e-6, 20
hmin, hmax = 1e-6, 3

x = np.linspace(xmin, xmax, gsize)
h = np.linspace(hmin, hmax, gsize)
X,H = np.meshgrid(x,h)


# ---- monopolist -------
fig1 = plt.figure()
plt.subplot(1, 2, 1)
U, V = mon.Dx(X,H), mon.Dh(X,H)  # x- and y-direction arrows size
M = np.hypot(U,V)  # re-scale arrows
U /= M
V /= M

plt.quiver( X, H, U, V, M, cmap = plt.cm.jet)
plt.title('Mon. - Phase plot')
plt.xlabel('Pop. $x$')
plt.ylabel('Harvest. $h$')

plt.plot(x, mon.g(x), linewidth=2.0)
plt.xlim([xmin, xmax])
plt.ylim([hmin, hmax])

fp_mon = [ 1.88160958,  0.98734758]  # fixed points
plt.plot(fp_mon[0], fp_mon[1], "red", marker = "o", markersize = 5.0)  # plot fixed points


# ---- social planner -------
plt.subplot(1, 2, 2)
U, V = soc.Dx(X,H), soc.Dh(X,H)  # x- and y-direction arrows size
M = np.hypot(U,V)  # re-scale arrows
U /= M
V /= M

plt.quiver( X, H, U, V, M, cmap = plt.cm.jet)
plt.title('Soc. Planner - Phase plot')
plt.xlabel('Pop. $x$')
plt.ylabel('Harvest. $h$')

plt.plot(x, mon.g(x), linewidth=2.0)
plt.xlim([xmin, xmax])
plt.ylim([hmin, hmax])

fp_soc = [1.74988984,  0.92744933]  # fixed points
plt.plot(fp_soc[0], fp_soc[1], "red", marker = "o", markersize = 5.0)  # plot fixed points

plt.tight_layout()
plt.savefig('mon_soc.pdf')
plt.show()
```

```python
# =========== phase-space plot ====================

def phase_space(flag, *pparameters):
    """Plot phase-space. Flag = 1 if monopolist, 0 if soc planner"""
    
    if flag==1:
        agent, atype = Monopolist(*parameters), 'Monopolist'
    elif flag==0:
        agent, atype = SocialPlanner(*parameters), 'Social Planner'
    
    gsize = 30
    xmin, xmax = 1e-6, 20
    hmin, hmax = 1e-6, 3
    x = np.linspace(xmin, xmax, gsize)
    h = np.linspace(hmin, hmax, gsize)
    X,H = np.meshgrid(x,h)
    
    U, V = agent.Dx(X,H), agent.Dh(X,H)  # x- and y-direction arrows size
    M = np.hypot(U,V)  # re-scale arrows
    U /= M
    V /= M
    
    agentfig = plt.figure()
    plt.title(atype + ' - Phase space plot')
    plt.xlabel('Pop. $x$')
    plt.ylabel('Harvest. $h$')
    plt.quiver( X, H, U, V, M, cmap = plt.cm.jet)
    
    plt.plot(x, agent.g(x), linewidth=2.0)
    plt.xlim([xmin, xmax])
    plt.ylim([hmin, hmax])
    
    return agentfig  # return plot with x-isocline
```

```python

myfig = phase_space(0, *parameters)
fp_soc = [1.74988984,  0.92744933]  # fixed points
plt.plot(fp_soc[0], fp_soc[1], "red", marker = "o", markersize = 10.0)  # plot fixed points
plt.show(myfig)
```
